```
CO2_emissions_df = pd.read_csv('API_EN.ATM.CO2E.KT_DS2_en_csv_v2_1345584.csv', index_col = 0, skiprows = 4, )
```

```
CO2_emissions_df.head()
```

```
CO2_emissions_df = CO2_emissions_df.drop(columns = ['Country Code', 'Indicator Name', 'Indicator Code', 'Unnamed: 64'])
```

```
CO2_emissions_df = CO2_emissions_df.transpose()

```

```
CO2_emissions_df
```

```
CO2_emissions_df.rename(columns = { 'Country Name' : 'Years'}, inplace = True) 
```

```
array2 = []
array2 = CO2_emissions_df.sum(axis=1)
array2
```

```
x=1960
array = []
for item in CO2_emissions_df['United States']:
  array.append(x)
  x=x+1

array

CO2_emissions_df['Years'] = array

```

```
CO2_emissions_df['Years']
```

```
data = { 'Years': array,
        'CO2_emissions': array2}

df_total_CO2 = pd.DataFrame(data, columns= ['Years', 'CO2_emissions'])
```

```
df_total_CO2
```

```
fires_and_co2_emissions = pd.merge(df_fires, df_total_CO2, on = 'Years', how='inner')
```

```
fires_and_co2_emissions
```

```
fires_and_co2_emissions['CO2_emissions'][30] = 319000000
fires_and_co2_emissions['CO2_emissions'][31] = 360000000
```

```
import numpy as np
fires_arr = np.array(fires_and_co2_emissions['Total Acres'])
co2_arr = np.array(fires_and_co2_emissions['CO2_emissions'])
corcef = np.corrcoef(fires_arr,co2_arr)
corcef
```

```
import matplotlib.pyplot as plt


fig, ax1 = plt.subplots() 
plt.title('Acres burned by Wildfires in CA and Worldwide CO2 emissions per year', fontname="Times New Roman",fontweight="bold", pad = 20)
ax1.plot(fires_and_co2_emissions['Years'],fires_and_co2_emissions['CO2_emissions'], color = 'black')
ax1.set_ylabel('Worldwide CO2 Emissions (Hundred Millions)', fontsize = 8, color = 'black')
ax1.ticklabel_format(useOffset=False, style='sci')
ax2 = ax1.twinx() 
  
ax2.plot(fires_and_co2_emissions['Years'],fires_and_co2_emissions['Total Acres'], color = 'firebrick') 
ax2.set_ylabel('Acres Burned from Wildfires in California (Millions)', fontsize = 8, color = 'black')
ax2.spines['right'].set_color('firebrick')
ax2.tick_params(axis='y', colors='firebrick')
ax2.ticklabel_format(useOffset=False, style='sci')
plt.show() 
```
